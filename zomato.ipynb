```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
# importing libraries
```

```python
#Reading data and getting first 5 rows 
df= pd.read_csv('D:\Python\zomato.csv')
df.head()
```

```python
#getting num of rows and columns
df.shape
```

```python
df.columns
```

```python
#dropping columns
df= df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'], axis=1)
df.head()
```

```python
df.info()
```

```python
#dropping duplicates
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
#removing /5 and string value
def handlerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value= str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#checking total number of null values
df.rate.isnull().sum()
```

```python
#taking mean to fill null values
df['rate'].fillna(df['rate'].mean(),inplace=True)
df.rate.isnull().sum()
```

```python
#dropping all null values
df.dropna(inplace=True)
df.head()
```

```python
#renaming columns
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'Type'}, inplace= True)
df.head()
```

```python
df.drop(['listed_in(city)'],axis=1)
```

```python
#handling comma
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace( ',', '' )
        return float(value)
    else:
        return float(value)

df['cost2plates']= df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
#making frequency as others for less than 1000
def handle_rest_type(value):
    if (value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()

    
```

```python
#cleaning location
location=df['location'].value_counts(ascending=False)
locationlessthan300=location[location<300]

def handlelocation(value):
    if(value in locationlessthan300):
        return 'others'
    else:
        return value
df['location']=df['location'].apply(handlelocation)
df['location'].value_counts()


```

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisineslessthan100=cuisines[cuisines<100]

def handlecuisines(value):
    if (value in cuisineslessthan100):
        return 'Others'
    else:
        return value

df['cuisines']=df['cuisines'].apply(handlecuisines)
df['cuisines'].value_counts()
```
